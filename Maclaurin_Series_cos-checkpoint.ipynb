```python
%matplotlib notebook
import math
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from matplotlib import animation
import random


fig, ax = plt.subplots(figsize = (5, 5))
ax.set_xlim(-10, 10)
ax.set_ylim(-4, 4)
ax.set_xlabel("x", fontsize = 15)
ax.set_ylabel("y", fontsize = 15)
ax.grid()

line, = ax.plot([0], [0], 'b')
x = np.linspace(-10, 10, 129)

def update(a):
    ans=0
    for i in range(int(a)):
        coef = (-1)**(i)
        num = x**(2*i)
        denom = math.factorial(2*i)
        ans += ( coef ) * ( (num)/(denom) )
        y=ans
        line.set_data(x, y)



ani = animation.FuncAnimation(
      fig,  # Figureã‚ªãƒ–ã‚¸ã‚§ã‚¯ãƒˆ
      update,  # ã‚°ãƒ©ãƒ•æç”»é–¢æ•°
      frames = 11,  # ãƒ•ãƒ¬ãƒ¼ãƒ ã‚’è¨­å®š
      interval = 1000,  # æ›´æ–°é–“éš”(ms)
      repeat = True,  # æç”»ã‚’ç¹°ã‚Šè¿”ã™
      blit = True  # blitting ã«ã‚ˆã‚‹å‡¦ç†ã®é«˜é€ŸåŒ–
      )

ani.save('ani_update.gif', writer='pillow')
plt.show()
```
