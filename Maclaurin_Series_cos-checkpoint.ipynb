```python
%matplotlib notebook
import math
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from matplotlib import animation
import random


fig, ax = plt.subplots(figsize = (5, 5))
ax.set_xlim(-10, 10)
ax.set_ylim(-4, 4)
ax.set_xlabel("x", fontsize = 15)
ax.set_ylabel("y", fontsize = 15)
ax.grid()

line, = ax.plot([0], [0], 'b')
x = np.linspace(-10, 10, 129)

def update(a):
    ans=0
    for i in range(int(a)):
        coef = (-1)**(i)
        num = x**(2*i)
        denom = math.factorial(2*i)
        ans += ( coef ) * ( (num)/(denom) )
        y=ans
        line.set_data(x, y)



ani = animation.FuncAnimation(
      fig,  # Figureオブジェクト
      update,  # グラフ描画関数
      frames = 11,  # フレームを設定
      interval = 1000,  # 更新間隔(ms)
      repeat = True,  # 描画を繰り返す
      blit = True  # blitting による処理の高速化
      )

ani.save('ani_update.gif', writer='pillow')
plt.show()
```
